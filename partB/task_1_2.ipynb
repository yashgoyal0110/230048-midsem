# Task 1.2 — Key Assumptions

**Paper:** *Learning from Distributions via Support Measure Machines*

## Assumption 1

**Assumption:** The embedding kernel k is *characteristic*, meaning the mean map μ: P → H is injective (different distributions produce different mean embeddings).

**Why the method needs it:** If two distinct distributions P ≠ Q have the same mean embedding μ_P = μ_Q, then the expected kernel K(P, Q) = K(P, P), and the SMM cannot distinguish them at all — the kernel would be blind to the distributional difference. The entire representational power of the SMM rests on injectivity.

**Violation scenario:** A linear kernel k(x,z) = ⟨x,z⟩ is not characteristic: it only captures first moments, so any two distributions with the same mean but different covariances (e.g., N(0,1) vs. N(0,4)) produce the same mean embedding. The SMM would classify them identically, regardless of their shapes.

**Paper reference:** Section 2, sentence following Equation (1) — "if k is characteristic, the map (1) is injective, i.e., all the information about the distribution is preserved [10]."

---

## Assumption 2

**Assumption:** Sufficient samples are available from each distribution P_i to estimate the empirical kernel K_emp accurately (Equation 4). The approximation error is O(m^{-1/2}).

**Why the method needs it:** The SMM's kernel K(P_i, P_j) is often unknown in closed form. The empirical estimate K_emp (Eq. 4) is used instead. If too few samples are available, the kernel matrix becomes noisy, degrading the SVM boundary and causing the SMM to behave no better than random.

**Violation scenario:** In a real-time sensor fusion application where each "distribution" is built from only 2–3 observations per time window, the empirical kernel will be dominated by sampling noise. The SMM would fail to capture any meaningful covariance signal and would collapse to near-chance performance.

**Paper reference:** Section 3.1, paragraph following Equation (4) — "A finite sample of size m from a distribution P suffices (with high probability) to compute an approximation within an error of O(m^{-1/2})."

---

## Assumption 3

**Assumption:** For the Flex-SVM equivalence (Lemma 4) to hold, the distributions P_i must differ from each other only in their *location* parameter (i.e., all P_i are shifts of a common base density g).

**Why the method needs it:** Lemma 4 derives a per-point kernel K_g by factoring the distribution as g(x_i, x), where x_i is the location. If distributions differ in shape, scale, or modality — not just location — then K_g cannot be factored this way, and the Flex-SVM reduction is invalid.

**Violation scenario:** In a natural scene classification task where each image's distribution has a completely different modality (unimodal histogram for a plain sky vs. multimodal for a cluttered indoor scene), the distributions cannot be characterized by a shared density template. The Flex-SVM would apply an incorrect kernel and the SMM would need to be used directly without this simplification.

**Paper reference:** Section 4.2, sentence after Lemma 4 — "The important assumption for this equivalence is that the distributions P_i differ only in their location in the parameter space."
